```python
import numpy as np
import pandas as pd
```

```python
autos_raw = pd.read_csv("autos.csv",encoding = "ISO-8859-1")
```

```python
autos = autos_raw[:2000][['price', 'yearOfRegistration', 'gearbox', 'powerPS', 'kilometer','fuelType', 'brand', 'vehicleType']]
#powerPS == metric HP

autos['price'] /= 100
#price in thousand euros

autos = autos.dropna(how='any')

autos = autos.drop(autos[autos.price < 1].index)
autos = autos.drop(autos[autos.price > 1000].index)

autos = autos.drop(autos[autos.powerPS == 0].index)
autos = autos.drop(autos[autos.powerPS > 1000].index)

autos = autos.drop(autos[autos.yearOfRegistration <1990].index)

n = autos.shape[0]
print(n)
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

x = 'powerPS'
y = 'price'
data = autos[[x, y]]

k = 3


k_means = KMeans(n_clusters=k).fit(data)
k_means_centers = k_means.cluster_centers_

labels = k_means.labels_


plt.title("K-means without standartization")
plt.grid()
plt.scatter(data[x], data[y], c = labels)
plt.xlabel(x)
plt.ylabel(y)
plt.scatter(k_means_centers[:,0], k_means_centers[:,1], c='r')
plt.show()

grand_mean = np.mean(k_means_centers, axis = 0)
print(grand_mean)
print(k_means_centers)
```

```python
def compute_mean_std(data):
    mean = np.mean(data)
    x_i_sq = data*data
    std = (np.mean(x_i_sq) - mean**2)**(1/2)
    return mean, std


#Standartization of dataset

mean_x, std_x = compute_mean_std(data[x])
mean_y, std_y = compute_mean_std(data[y])

new_data = np.copy(data)

new_data[:,0] = (data[x]-mean_x)/std_x
new_data[:,1] = (data[y]-mean_y)/std_y

k_means = KMeans(n_clusters=k).fit(new_data)
k_means_centers = k_means.cluster_centers_

labels = k_means.labels_


plt.title("K-means with standartization")
plt.grid()
plt.scatter(new_data[:,0], new_data[:,1], c = labels)
plt.xlabel(x)
plt.ylabel(y)
plt.plot(k_means_centers[:,0], k_means_centers[:,1], 'r*')
plt.show()


grand_mean = np.mean(k_means_centers, axis = 0)
print(grand_mean)
print(k_means_centers)
```

```python
def bootstrap(data, m = 5000):
    m = 5000
    n = data.shape[0]
    r = np.random.randint(0,n,(n,m))
    xr = data[r]
    mr = np.mean(xr, axis = 0)
    return mr

mr = bootstrap(np.array(data["price"][labels==0]))
mean_x, std_x = compute_mean_std(data['price'][labels==0])
print("Cluster mean =",mean_x)
print("Cluster std  =",std_x)

plt.title("Histogram of price values in first cluster")
plt.xlabel('Value')
plt.grid()
plt.ylabel('Counts')
plt.hist(data['price'][labels==0])
plt.show()


plt.title("Bootstrap of price values in first cluster")
plt.xlabel('Value')
plt.grid()
plt.ylabel('Counts')
plt.hist(mr)
plt.show()
```

```python
# assignment 5

columns = ['price', 'yearOfRegistration', 'powerPS']

data = autos[columns]

data_ranges = np.copy(data)
data_std = np.copy(data)

data_ranges = (data_ranges - data_ranges.mean()) / (data_ranges.max() - data_ranges.min())

for i, name in enumerate(columns):
    mean, std = compute_mean_std(data[name])
    data_std[:,i] = (data[name] - mean) / std

# print(np.linalg.svd(data_ranges)[0])
# print(np.linalg.svd(data_std)[0])

svd_ranges = np.linalg.svd(data_ranges)[0]
svd_std = np.linalg.svd(data_std)[0]

plt.title("SVD with ranges-based normalization")
plt.grid()
plt.scatter(svd_ranges[:,0], svd_ranges[:,1], c = labels)
plt.show()

plt.title("SVD with stddev-based normalization")
plt.grid()
plt.scatter(svd_std[:,0], svd_std[:,1], c = labels)
plt.show()
```
